```python
import pyautogui
import subprocess
import time
import pandas as pd
import smtplib
import email.message
import os
from dotenv import load_dotenv, find_dotenv

# O find_dotenv() vasculha a sua pasta e acha o arquivo automaticamente
load_dotenv(find_dotenv())

# PASSO 1: REQUISIÇÃO DE ABERTURA - SISTEMA DA EMPRESA


chrome_path = r'C:\Program Files\Google\Chrome\Application\chrome.exe'
profile_folder = "Daniel"
Link = 'https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga'

def open_specific_profile():
    # Link diretamente no comando
    command = f'"{chrome_path}" --profile-directory="{profile_folder}" "{Link}"'
    
    try:
        subprocess.Popen(command)
        print(f"Abrindo o Chrome no perfil '{profile_folder}' direto no galpão (Drive)...")
        time.sleep(5) # Tempo para a página carregar
        pyautogui.hotkey('win', 'up') # Maximiza a janela
        
    except FileNotFoundError:
        print("Chrome não encontrado no caminho especificado.")

open_specific_profile()



# PASSOS 2 E 3: NAVEGAÇÃO E EXPORTAÇÃO COM TRAVA DE SEGURANÇA


time.sleep(2)
print("\nProcurando a pasta 'Exportar' na tela...")

download_concluido = False 

try:
    posicao_exportar = pyautogui.locateCenterOnScreen('botao_exportar.png', confidence=0.8, grayscale=True)
    
    if posicao_exportar is not None:
        pyautogui.moveTo(posicao_exportar, duration=0.5) 
        pyautogui.doubleClick() 
        print("✅ Duplo clique na pasta Exportar! (Abrindo a pasta...)")
        
        time.sleep(4) # Espera a pasta abrir e carregar
        
        # --- NOVO PASSO: SELECIONAR O ARQUIVO ---
        print("Procurando o arquivo Excel para selecionar...")
        posicao_arquivo = pyautogui.locateCenterOnScreen('arquivo_excel.png', confidence=0.8, grayscale=True)
        
        if posicao_arquivo is not None:
            pyautogui.moveTo(posicao_arquivo, duration=0.5)
            pyautogui.click() # Clique SIMPLES para selecionar
            print("✅ Arquivo selecionado! O botão de baixar deve aparecer agora.")
            
            time.sleep(1.5) # Dá um tempinho para a animação do Google Drive mostrar o botão
            
            # --- AGORA SIM PROCURA O BOTÃO BAIXAR ---
            print("Procurando o botão 'Baixar' na tela...")
            posicao_baixar = pyautogui.locateCenterOnScreen('botao_baixar.png', confidence=0.8, grayscale=True)
            
            if posicao_baixar is not None:
                pyautogui.moveTo(posicao_baixar, duration=0.5)
                pyautogui.click()
                print("Iniciando o download...")
                
                print("Aguardando 7 segundos para o arquivo chegar na pasta...")
                time.sleep(7) 
                
                download_concluido = True 
                
            else:
                print("Erro: Não achei o botão 'Baixar' mesmo após selecionar o arquivo.")
        else:
            print("Erro: Não achei o arquivo Excel dentro da pasta.")
            
    else:
        print("Erro: Não achei a pasta 'Exportar'.")

except Exception as e:
    print(f"Aviso do sistema de visão: O robô não encontrou a imagem esperada na tela. Detalhe: {e}")



# PASSO 4: ANÁLISE DE DADOS (CONDICIONADA AO DOWNLOAD)


# O código só entra aqui se a etapa clicar funcionar ou seja o robô localizar o 'Baixar'
if download_concluido:
    print("\nIniciando a leitura de dados com o Pandas...")
    caminho = r"C:\Users\danie\Downloads\Vendas - Dez.xlsx"

    try:
        tabela = pd.read_excel(caminho)
        display(tabela)

        faturamento_total = tabela["Valor Final"].sum()
        print(f"\n Faturamento Total: R${faturamento_total:,.2f}")

        qtde_produtos = tabela["Quantidade"].sum()
        print(f"Quantidade Total: {qtde_produtos}")  

        vendas_por_produto = tabela.groupby("Produto")["Valor Final"].sum().reset_index()
        print("\n Vendas por Produto:")
        display(vendas_por_produto)

        vendas_por_mes = tabela.groupby("Data")["Valor Final"].sum().reset_index()  
        print("\n Vendas por Mês:")
        display(vendas_por_mes)
        
    except FileNotFoundError:
        print(f"\n Erro no Pandas: O arquivo não estava em {caminho}.")
else:
    print("\n Operação Cancelada: A análise de dados não foi iniciada porque houve falha no processo de download.")
```

```python
# ==============================================================================
# PASSO 5: ENVIAR O E-MAIL PELO BACK-END (SMTP)
# ==============================================================================

print("\nMontando o e-mail com os resultados...")

meu_email = os.getenv("meu_email")
senha_app = os.getenv("senha_app")
email_destino = os.getenv("email_destino")

# 2. Montando o "Envelope"
# ==============================================================================
# PASSO 5: ENVIAR O E-MAIL PELO BACK-END (SMTP)
# ==============================================================================

print("\nMontando o e-mail com os resultados...")

meu_email = os.getenv("meu_email")
senha_app = os.getenv("senha_app")
email_destino = os.getenv("email_destino")

# 2. Montando o "Envelope"
msg = email.message.EmailMessage()
msg['Subject'] = "Relatório Automático de Vendas - Dezembro"
msg['From'] = meu_email
msg['To'] = email_destino

# 3. Escrevendo a "Carta"
# Usamos o 'f' antes das aspas para o Python entender que vamos injetar variáveis ali dentro
# Verificar se as variáveis foram definidas (caso o download não tenha sido concluído)
if 'faturamento_total' not in globals():
    faturamento_total = 0
    qtde_produtos = 0

corpo_email = f"""
Olá! Tudo bem?

Segue o resumo dos indicadores de vendas de Dezembro, processados automaticamente:

Faturamento Total: R$ {faturamento_total:,.2f}
Quantidade de Produtos: {qtde_produtos}

Este é um e-mail gerado de forma 100% autônoma pelo Python.

Atenciosamente,
Daniel Aleixo
"""
msg.set_content(corpo_email)

# 4. Entregando aos Correios (Conexão com o Servidor do Google)
try:
    print("Conectando ao servidor do Gmail...")
    # Abre a conexão segura na porta 465 (Padrão do Gmail)
    s = smtplib.SMTP_SSL('smtp.gmail.com', 465) 
    
    # Faz o login com a sua Senha de App
    s.login(meu_email, senha_app) 
    
    # Envia o envelope
    s.send_message(msg) 
    
    # Fecha a porta do servidor (Sempre feche a conexão!)
    s.quit() 
    print("E-mail enviado com sucesso! Olhe a sua caixa de entrada!")
    
except Exception as e:
    print(f"Erro na conexão com o servidor de e-mail. Detalhe: {e}")
```
